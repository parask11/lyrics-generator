```python
import numpy as np
#np.random.seed(11)
```

```python
text_path = "Apna Time Aayega.txt"
```

```python
def load_text(filename):
    with open(filename,encoding='utf-8') as f:
        return f.read().lower()

text = load_text(text_path)
```

```python
print(text)
```

```python
def generateTable(data,k=4):
    T = {}
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        else:
            if T[X].get(Y) is None:
                T[X][Y] = 1
            else:
                T[X][Y] += 1
                
    for kx in T.keys():
        s = float(sum(T[kx].values()))
        for k in T[kx].keys():
            T[kx][k] = T[kx][k]/s
        
    return T
```

```python
model = generateTable(text)
```

```python
model
```

```python
def sample_next(ctx,T,k):
    #np.random.seed(11)
    ctx = ctx[-k:]
    if T.get(ctx) is None:
        return " "
    
    possible_chars = list(T[ctx].keys())
    possible_values = list(T[ctx].values())
    #print(possible_chars,possible_values)
    
    return np.random.choice(possible_chars,p=possible_values)
```

```python
def generateLyrics(starting_sent,k=4,maxlen=2000):
    sentence = starting_sent
    ctx = sentence[-k:]
    for i in range(maxlen):
        next_char = sample_next(ctx,model,k)
        sentence += next_char
        ctx = sentence[-k:]
    return sentence
```

```python
np.random.seed(11)
new = generateLyrics("apna")
```
